```python
import pandas as pd
df = pd.read_json("../tm_real_samples.json")
```

```python
df.head()
```

```python
# remove target variable
df.drop('target', axis=1, inplace=True)
```

```python
# turn days back into dates
import numpy as np
date_cols = ['validity_start', 'validity_end', 'date_of_birth']
for col in date_cols:
        df[col] = pd.to_datetime(np.floor(df[col]), unit='D', origin='1970-01-01').dt.strftime('%Y-%m-%d')
```

```python
# turn float values to int
float_to_int_cols = ["age", "season_id", "league_played_matches", "league_minutes_played", "league_goals", "international_played_matches", "international_minutes_played", "international_goals", "missed_matches"]
df[float_to_int_cols] = df[float_to_int_cols].apply(np.floor).astype('Int64')
```

```python
# store post proccessed stuff
df.to_json("../tm_postprocessed_excerpt.json", orient="records")
```
